```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

<h1 style="color: #DE006F">CATEGORICAL DATA</h1>

<h3 style="color: red">Using Countplot- for categorical data, we try to finf the frequency count for each category</h3>

<h4>For 'survived':</h4>

```python
sns.countplot(x='Survived',
             data=df,
             color='pink')
```

<h4>For sex:</h4>

```python
sns.countplot(x='Sex',
           data=df,
           color='pink')
```

<h4>For 'Pclass':</h4>

```python
sns.countplot(x='Pclass',
             data=df,
             color='pink')
```

<h3 style="color: red">Using Piechart- for categorical data, we try to find the % the frequency count for each category</h3>

<h4>For 'survived':</h4>

```python
#Seaborn does not have a built-in method for creating pie charts directly. 
survived_count=df['Survived'].value_counts()
survived_count
plt.pie(survived_count,
        labels=survived_count.index,
        autopct='%2.f%%',
        colors=['pink','purple'])
plt.show()
```

<h4>For sex:</h4>

```python
sex_count=df['Sex'].value_counts()
sex_count
plt.pie(sex_count,
        labels=sex_count.index,
        autopct='%2.f%%',
        colors=['pink','purple'])
plt.show()
```

<h4>For Pclass:</h4>

```python
pclass_counts=df['Pclass'].value_counts()
pclass_counts

plt.pie(pclass_counts,
       labels=pclass_counts.index,
       autopct='%2.f%%',
       colors=['pink','purple'])
plt.show()
```

<h3 style="color: red">Using Histogram- it is used to visualize the distribution of a continuous numerical variable </h3> 

<h4>For 'Age':</h4>

```python
plt.hist(df['Age'],
        color='#DE006F',
        bins=10)
plt.show()
```

<h3 style="color: red">Using Distplot- used to visualize the distribution of a numerical variable, combining a histogram and a Kernel Density Estimate (KDE) curve for a smoother representation.</h3> 

<h4>For 'Age':</h4>

```python
sns.distplot(df['Age'],
            color='#DE006F')
```

<h3 style="color: red">Using Boxplot- used to visualize the distribution, central tendency, and variability of a dataset, highlighting outliers and the spread of data through quartiles. GIVES A 5 NUMBER SUMMARY</h3> 

<h4>For 'Age':</h4>

```python
sns.boxplot(df['Age'],
           color="pink")
```

<h4>For 'Fare':</h4>

```python
sns.boxplot(df['Fare'],
           color='pink')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h1 style='color:#DE006F'>LOADING THE DATASETS</h1>

<h3 style='color:red'>TIPS DATASET</h3>

```python
tips=sns.load_dataset('tips')
tips.head()
```

<h3 style='color:red'>TITANIC DATASET</h3>

```python
titanic=pd.read_csv('Titanic-Dataset.csv')
titanic.head(3)
```

<h3 style='color:red'>FLIGHTS DATASET</h3>

```python
flights=sns.load_dataset('flights')
flights.head()
```

<h3 style='color:red'>IRIS DATASET</h3>

```python
iris=sns.load_dataset('iris')
iris.head()
```

<h1 style='color:#DE006F'> 1. Scatterplot -(Numerical-Numerical) </h1>

<h3 style='color:red'> BIVARIANT ANALYSIS </h3>

```python
#finding a relationship between tips and total bill
sns.scatterplot(x=tips['tip'],
               y=tips['total_bill'],
               color='#DE006F')
#we can see that there is a linear sort of a relationsip between bill and tip
```

<h3 style='color:red'>MULTIVARIANT ANALYSIS </h3>

```python
#The code creates a scatter plot showing the relationship between total bill and tip amounts in the tips dataset, with points colored by the sex column.

sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
#Thus, we are getting 3 pieces of information
```

```python
#The code creates a scatter plot showing the relationship between total bill and tip amounts in the tips dataset, 
#with points colored by size and styled by the smoker status.

sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['size'],style=tips['smoker'])
#Thus, we are getting 4 pieces of information
```

```python
tips.head(3)
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['time'])

#Thus, we are getting 5 pieces of information
```

<h2 style='color:#DE006F'> Barplot (Numerical-Categorical) </h2>

 <h3 style='color:red'>Bivariant Analysis</h3>

```python
titanic.head(3)
```

```python
sns.barplot(y=titanic['Pclass'],x=titanic['Embarked'])
# The black lines that we get on top of each bar is "confidence interval"
```

<h3 style='color:red'>Multivariant Analysis </h3>

<h1 style='color:#DE006F'> 3. Box Plot(Numerical-Categorical) </h1>

<h3 style='color:red'>BIVARIANT ANALYSIS </h3>

```python
titanic.head(2)
```

```python
sns.boxplot(x=titanic['Sex'],
            y=titanic['Age'],
           palette='pink')
plt.show()
```

<h3 style='color:red'>MULTIVARIANT ANALYSIS </h3>

```python
sns.boxplot(x=titanic['Sex'],
           y=titanic['Age'],
           hue=titanic['Survived'],
           palette='pink')
```

<h1 style='color:#DE006F'>4. Distplot (Numerical-Categorical) </h1>

<h3 style='color:red'>BIVARIANT ANALYSIS </h3>

```python
## displot of people who survived and people who did not
sns.displot(titanic['Age'],
            kde=True)
```

<h3 style='color:red'>MULTIVARIANT ANALYSIS </h3>

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

 <h1 style='color:#DE006F'>5. HeatMap( Categorical-Categorical) </h1>

<h3 style='color:red'>BIVARIANT ANALYSIS </h3>

```python
#EX: To find out how many died and how many survived in Pclass
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
#To  creat a heatmap
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# titanic.groupby('Pclass')['Survived'].mean()
titanic.groupby('Pclass')['Survived'].mean()*100  # by multiplying by 100,it gives the percentage
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

<h3 style='color:red'>MULTIVARIANT ANALYSIS </h3>

<h1 style='color:#DE006F'>6. ClusterMap (Categorical-Categorical) </h1>

```python
# creating a crosstab
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
# plotting the crosstab   dendograms
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

<h1 style='color:#DE006F'>8. Lineplot (Numerical -Numerical) </h1>

```python
#lineplot should be used when quantity in x axis should be time based

flights.head()
```

```python
# Ensure that only numeric columns are being summed
#year wise how many passengers 'flew'

new=flights.groupby('year').sum(numeric_only=True).reset_index()
new
```

```python
#PLOTTING IT:

#sns.lineplot(x=new['year'],y=new['passengers'])
sns.lineplot(x=new['year'],y=new['passengers'])
```
